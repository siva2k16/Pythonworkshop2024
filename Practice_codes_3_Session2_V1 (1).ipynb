```python
# Base class
class Animal:
    def __init__(self, name):
        self.name = name  # Attribute

    def speak(self):     # Abstract method
        raise NotImplementedError("Subclass must implement abstract method")

# Derived class (inherits from Animal)
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Another derived class
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

# Using Encapsulation to access the data
my_dog = Dog("Buddy")
print(my_dog.speak())  # Buddy says Woof!

my_cat = Cat("Kitty")
print(my_cat.speak())  # Kitty says Meow!
```

```python
import cv2
from google.colab.patches import cv2_imshow

class ImageProcessor:
    def __init__(self, image_path):
        # Load the image from a file
        self.image = cv2.imread(image_path)
        if self.image is None:
            raise ValueError("Image not found or the path is incorrect")

    def createdata(self, image_path):
        self.image = cv2.imread(image_path)

    def display_image(self, window_name="Image"):
        # Display the image in a window
        cv2_imshow(self.image)
        #cv2.destroyAllWindows()

    def convert_to_grayscale(self):
        # Convert the image to grayscale
        gray_image = cv2.cvtColor(self.image, cv2.COLOR_BGR2GRAY)
        return gray_image

# Example usage of the class
image_path = '/content/sample_data/demo1.png'  # Ensure you have an image at this path
processor = ImageProcessor(image_path)
#processor = ImageProcessor()
#processor.createdata(image_path)
processor.display_image("Original Image")
gray_image = processor.convert_to_grayscale()
cv2_imshow(gray_image)
cv2.destroyAllWindows()
```

```python
class MyClass:
    public_class_attribute = 'I am a public class attribute'
    _protected_class_attribute = 'I am a protected class attribute'
    __private_class_attribute = 'I am a private class attribute'

    def __init__(self):
        self.public_instance_attribute = 'I am a public instance attribute'
        self._protected_instance_attribute = 'I am a protected instance attribute'
        self.__private_instance_attribute = 'I am a private instance attribute'

    def get_private_instance_attribute(self):
        return self.__private_instance_attribute

# Access examples
obj = MyClass()

# Accessing public attributes
print(obj.public_instance_attribute)  # Accessible
print(MyClass.public_class_attribute)  # Accessible

# Accessing protected attributes (by convention should not be accessed directly)
print(obj._protected_instance_attribute)  # Possible, but against convention
print(MyClass._protected_class_attribute)  # Possible, but against convention

# Accessing private attributes - requires special syntax since they are name-mangled
print(obj.get_private_instance_attribute())  # Possible using a method
print(obj._MyClass__private_instance_attribute)  # Accessing name-mangled attribute directly
```

```python
#Basic Code (Using List, Set, Tuple, Dictionary)
# Synthetic Retail Data
products = [
    ("T-shirt", 19.99),
    ("Jeans", 39.99),
    ("Sneakers", 59.99),
    ("Hat", 12.99)
]

# Set of discounted products
discounted_products = {"Jeans", "Hat"}

# Dictionary to hold the product discounts
discounts = {
    "Jeans": 0.2,  # 20% discount
    "Hat": 0.1     # 10% discount
}

# List to hold discounted prices
discounted_prices = []

# Loop to calculate discounted prices
for product in products:
    name, price = product
    if name in discounted_products:
        discount = discounts[name]
        discounted_price = price * (1 - discount)
        discounted_prices.append((name, discounted_price))
    else:
        discounted_prices.append(product)

print(discounted_prices)
```

```python
#Code With Functions
def apply_discounts(products, discounted_products, discounts):
    discounted_prices = []
    for product in products:
        name, price = product
        if name in discounted_products:
            discount = discounts[name]
            discounted_price = price * (1 - discount)
            discounted_prices.append((name, discounted_price))
        else:
            discounted_prices.append(product)
    return discounted_prices

# Usage of the function
discounted_prices = apply_discounts(products, discounted_products, discounts)
print(discounted_prices)
```

```python
#Code With OOP
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def apply_discount(self, discount):
        self.price *= (1 - discount)

class RetailStore:
    def __init__(self, products):
        self.products = products
        self.discounts = {}

    def add_discount(self, product_name, discount):
        self.discounts[product_name] = discount

    def apply_discounts(self):
        for product in self.products:
            if product.name in self.discounts:
                product.apply_discount(self.discounts[product.name])

# Usage of the classes
store = RetailStore([Product("T-shirt", 19.99), Product("Jeans", 39.99), Product("Sneakers", 59.99), Product("Hat", 12.99)])
store.add_discount("Jeans", 0.2)
store.add_discount("Hat", 0.1)
store.apply_discounts()

# Display discounted prices
for product in store.products:
    print(f"Product: {product.name}, Price: ${product.price}")
```
